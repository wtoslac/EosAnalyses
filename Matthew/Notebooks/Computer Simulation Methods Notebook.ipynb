```python
import random
import matplotlib.pyplot as plt

class agent():

    def __init__(self, money) -> None:
        self.money = money
    
m0 = 1000
iterations = 10000
i = 0
N = 1000
```

```python
agents = []
for i in range(N):
    agents.append(agent(m0))
while i < iterations:
    agent1 = agents[int(random.uniform(0,N-1))]
    agent2 = agents[int(random.uniform(0,N-1))]

    agent1.money = random.uniform(0,1)*(agent1.money + agent2.money)
    agent2.money = (1-random.uniform(0,1))*(agent1.money + agent2.money)
    i += 1

total_money = []
for person in agents:
    total_money.append(person.money)

plt.hist(total_money, bins=50)
plt.show()
```

```python
saveings_Model = 0.25
agents = []
for i in range(N):
    agents.append(agent(m0))
i = 0

while i < iterations:

    espislon = random.uniform(0,1)

    agent1 = agents[int(random.uniform(0,N-1))]
    agent2 = agents[int(random.uniform(0,N-1))]

    deltam = (1-saveings_Model)*(espislon*agent2.money - (1-espislon)*agent1.money)
    # print("Agent1:", agent1.money)
    # print("Agent2:", agent2.money)
    # print("Epislon:",espislon)
    # print("Deltam:",deltam)

    agent1.money = agent1.money + deltam
    agent2.money = agent2.money - deltam
    i += 1

total_money = []
for person in agents:
    total_money.append(person.money)

#print(total_money)

plt.hist(total_money,bins=50)
plt.show()
```
